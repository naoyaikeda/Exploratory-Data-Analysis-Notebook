```python
import numpy as np
import pandas as pd
 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
german_df = pd.read_csv('Data/german.csv', index_col=0)
```

```python
german_df['Class'] = german_df['Class'].astype('category')
```

```python
sns.pairplot(german_df, hue = 'Class', vars = ['Duration in month', 'Credit amount'])
```

```python
sns.pairplot(german_df, hue = 'Class', vars = ['Installment rate in percentage of disposable income', 'Age in years',])
```

```python
sns.pairplot(german_df, hue = 'Class', vars = ['Number of existing credits at this bank', 'Number of people being liable to provide maintenance for'])
```
